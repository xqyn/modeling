```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative

# Define polynomial functions
def linear(x):
    """f(x) = 2x + 1"""
    return 2*x + 1

def quadratic(x):
    """f(x) = x^2 - 3x + 2"""
    return x**2 - 3*x + 2

def cubic(x):
    """f(x) = x^3 - 2x^2 + x + 1"""
    return x**3 - 2*x**2 + x + 1

def quartic(x):
    """f(x) = 0.1x^4 - x^3 + 2x^2 - x + 1"""
    return 0.1*x**4 - x**3 + 2*x**2 - x + 1

# Generate x values
x = np.linspace(-5, 5, 500)

# Store functions and their names
functions = [
    (linear, "Linear: f(x) = 2x + 1"),
    (quadratic, "Quadratic: f(x) = x² - 3x + 2"),
    (cubic, "Cubic: f(x) = x³ - 2x² + x + 1"),
    (quartic, "Quartic: f(x) = 0.1x⁴ - x³ + 2x² - x + 1")
]

# Create subplots
fig, axes = plt.subplots(4, 3, figsize=(15, 12))
fig.suptitle('Polynomial Functions and Their Derivatives', fontsize=16, fontweight='bold')

for i, (func, title) in enumerate(functions):
    # Calculate function values
    y = func(x)

    # Calculate first derivative numerically
    dy = np.array([derivative(func, xi, dx=1e-6, n=1) for xi in x])

    # Calculate second derivative numerically
    d2y = np.array([derivative(func, xi, dx=1e-6, n=2) for xi in x])

    # Plot original function
    axes[i, 0].plot(x, y, 'b-', linewidth=2)
    axes[i, 0].axhline(y=0, color='k', linestyle='--', alpha=0.3)
    axes[i, 0].axvline(x=0, color='k', linestyle='--', alpha=0.3)
    axes[i, 0].set_title(f'{title}')
    axes[i, 0].set_xlabel('x')
    axes[i, 0].set_ylabel('f(x)')
    axes[i, 0].grid(True, alpha=0.3)

    # Plot first derivative
    axes[i, 1].plot(x, dy, 'r-', linewidth=2)
    axes[i, 1].axhline(y=0, color='k', linestyle='--', alpha=0.3)
    axes[i, 1].axvline(x=0, color='k', linestyle='--', alpha=0.3)
    axes[i, 1].set_title("First Derivative: f'(x)")
    axes[i, 1].set_xlabel('x')
    axes[i, 1].set_ylabel("f'(x)")
    axes[i, 1].grid(True, alpha=0.3)

    # Plot second derivative
    axes[i, 2].plot(x, d2y, 'g-', linewidth=2)
    axes[i, 2].axhline(y=0, color='k', linestyle='--', alpha=0.3)
    axes[i, 2].axvline(x=0, color='k', linestyle='--', alpha=0.3)
    axes[i, 2].set_title("Second Derivative: f''(x)")
    axes[i, 2].set_xlabel('x')
    axes[i, 2].set_ylabel("f''(x)")
    axes[i, 2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print analytical derivatives for reference
print("Analytical Derivatives:")
print("\n1. Linear: f(x) = 2x + 1")
print("   f'(x) = 2")
print("   f''(x) = 0")

print("\n2. Quadratic: f(x) = x² - 3x + 2")
print("   f'(x) = 2x - 3")
print("   f''(x) = 2")

print("\n3. Cubic: f(x) = x³ - 2x² + x + 1")
print("   f'(x) = 3x² - 4x + 1")
print("   f''(x) = 6x - 4")

print("\n4. Quartic: f(x) = 0.1x⁴ - x³ + 2x² - x + 1")
print("   f'(x) = 0.4x³ - 3x² + 4x - 1")
print("   f''(x) = 1.2x² - 6x + 4")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define polynomial functions
def linear(x):
    """f(x) = 2x + 1"""
    return 2*x + 1

def quadratic(x):
    """f(x) = x^2 - 3x + 2"""
    return x**2 - 3*x + 2

def cubic(x):
    """f(x) = x^3 - 2x^2 + x + 1"""
    return x**3 - 2*x**2 + x + 1

def quartic(x):
    """f(x) = 0.1x^4 - x^3 + 2x^2 - x + 1"""
    return 0.1*x**4 - x**3 + 2*x**2 - x + 1

# Generate x values
x = np.linspace(-5, 5, 500)

# Store functions and their names
functions = [
    (linear, "Linear: f(x) = 2x + 1"),
    (quadratic, "Quadratic: f(x) = x² - 3x + 2"),
    (cubic, "Cubic: f(x) = x³ - 2x² + x + 1"),
    (quartic, "Quartic: f(x) = 0.1x⁴ - x³ + 2x² - x + 1")
]

# Create subplots
fig, axes = plt.subplots(4, 3, figsize=(15, 12))
fig.suptitle('Polynomial Functions and Their Derivatives', fontsize=16, fontweight='bold')

for i, (func, title) in enumerate(functions):
    # Calculate function values
    y = func(x)

    # Calculate first derivative using numpy gradient
    dy = np.gradient(y, x)

    # Calculate second derivative
    d2y = np.gradient(dy, x)

    # Plot original function
    axes[i, 0].plot(x, y, 'b-', linewidth=2)
    axes[i, 0].axhline(y=0, color='k', linestyle='--', alpha=0.3)
    axes[i, 0].axvline(x=0, color='k', linestyle='--', alpha=0.3)
    axes[i, 0].set_title(f'{title}')
    axes[i, 0].set_xlabel('x')
    axes[i, 0].set_ylabel('f(x)')
    axes[i, 0].grid(True, alpha=0.3)

    # Plot first derivative
    axes[i, 1].plot(x, dy, 'r-', linewidth=2)
    axes[i, 1].axhline(y=0, color='k', linestyle='--', alpha=0.3)
    axes[i, 1].axvline(x=0, color='k', linestyle='--', alpha=0.3)
    axes[i, 1].set_title("First Derivative: f'(x)")
    axes[i, 1].set_xlabel('x')
    axes[i, 1].set_ylabel("f'(x)")
    axes[i, 1].grid(True, alpha=0.3)

    # Plot second derivative
    axes[i, 2].plot(x, d2y, 'g-', linewidth=2)
    axes[i, 2].axhline(y=0, color='k', linestyle='--', alpha=0.3)
    axes[i, 2].axvline(x=0, color='k', linestyle='--', alpha=0.3)
    axes[i, 2].set_title("Second Derivative: f''(x)")
    axes[i, 2].set_xlabel('x')
    axes[i, 2].set_ylabel("f''(x)")
    axes[i, 2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print analytical derivatives for reference
print("Analytical Derivatives:")
print("\n1. Linear: f(x) = 2x + 1")
print("   f'(x) = 2")
print("   f''(x) = 0")

print("\n2. Quadratic: f(x) = x² - 3x + 2")
print("   f'(x) = 2x - 3")
print("   f''(x) = 2")

print("\n3. Cubic: f(x) = x³ - 2x² + x + 1")
print("   f'(x) = 3x² - 4x + 1")
print("   f''(x) = 6x - 4")

print("\n4. Quartic: f(x) = 0.1x⁴ - x³ + 2x² - x + 1")
print("   f'(x) = 0.4x³ - 3x² + 4x - 1")
print("   f''(x) = 1.2x² - 6x + 4")
```
